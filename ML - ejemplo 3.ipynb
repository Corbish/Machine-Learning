```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline



###########################################################################################################
#                                     Adquisición de datos
#En este apartado se cargan los datos obtenidos de http://archive.ics.uci.edu/ml/datasets/Computer+Hardware,
#y se los coloca en un dataframe. Se realiza una copia de seguridad por cualquier modificación que sufra el dataframe
############################################################################################################


data = open('machine.data','r')

df = pd.read_csv(data, sep=",")
data.close()
df1 = df.copy(deep=True) #Mantengo una copia de seguridad 





##############################################################################################################
#                                           Procesamiento
#El procesamiento consiste en armar el dataframe para lo que convenga. La base de datos tiene 9 columnas
#la última corresponde a valores predecidos por regresión por otros programadores. En este caso se visualizan
#las primeras 8 columnas. Las dos primeras son marca y modelo
##############################################################################################################

#Doy valores numéricos a las características no predictivas (marca y modelo)
le = LabelEncoder()
df.iloc[:,0] = le.fit_transform(df.iloc[:,0]) 

le1 = LabelEncoder()
df.iloc[:,1] = le1.fit_transform(df.iloc[:,1]) 


#Visualización de correlación de datos: se verán las características que muestren mayor tendencia lineal
#

#sns.pairplot(df.iloc[:,0:-1],  corner=True)
        
##########################################################################################################
#                          Elección de datos test y train
#En la elección se considera los atributos que se visualizan con alta tendencia lineal. 
#Se observa que dichos atributos corresponden a la columna 3 del dataframe 
#Puesto que las particiones se harán de diferentes maneras, se prueban algunos valores aleatoriamente para
#saber cuál es el más acertado
##########################################################################################################
X = df.iloc[:,3:8]
y = df.iloc[:,8]



save = []
maxim = 0.5
#Se realizan diferentes calculos con diferentes tamaños de test y eligiendo datos de forma aleatorea. Como resultado se 
#obtiene que el tamaño adecuado de test es del30%
for i in range(1,4):
    for j in range(10):
        xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = i*0.1, random_state = j*50, shuffle = True)
        reg = linear_model.LinearRegression()
        reg.fit(xTrain, yTrain)
        yPred = pd.DataFrame(data = reg.predict(xTest))
        r2 = r2_score(yTest, yPred)
        print(r2)
        if r2 >= maxim:
            maxim = r2
            save = [i,j,r2]
            print(save)
#Se usan los valores obtenidos para el mayor r2, en base a eso se establece la regresión lineal correspondiente
xTrain, xTest, yTrain, yTest = train_test_split(X, y, 
                                                test_size = save[0]*0.1, random_state = save[1]*20, shuffle = True)
reg = linear_model.LinearRegression()
reg.fit(xTrain, yTrain)
yPred = pd.DataFrame(data = reg.predict(xTest))
r2 = r2_score(yTest, yPred)
print("El valor de regresión lineal es de {0:.3f}".format(r2))
```
